```julia
using DelimitedFiles
using Plots
data_arr = readdlm("C:/Users/Abhishek/Documents/hw1a.txt")
```

```julia
function solfirst(k)
    sum = 0.0
    for i in 1:k
        sum+=data_arr[i]
    end
    sum/k
end
```

```julia
y = rand(50)
x = 1:50
```

```julia

for i in 1:50
    y[i] = solfirst(10+10*(i-1))
end
p1 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=100:10:150)
for i in 1:50
    y[i] = solfirst(20+20*(i-1))
end
p2 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=100:10:150)
for i in 1:50
    y[i] = solfirst(50+50*(i-1))
end
p3 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=100:10:150)
for i in 1:50
    y[i] = solfirst(100+100*(i-1))
end
p4 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=100:10:150)
for i in 1:50
    y[i] = solfirst(200+200*(i-1))
end
p5 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=100:10:150)
p6 = plot(1:10,rand(10))
plot(p1,p2,p3,p4,p5,p6layout=(2,3),legend=false)
```

```julia
function solsecond(k)
    x = rand(1:(length(data_arr)-k))
    sum = 0.0
    for i in x:x+k
        sum+=data_arr[i]
    end
    sum/k
end 
```

```julia

for i in 1:50
    y[i] = solsecond(10)
end
p1 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:150)
for i in 1:50
    y[i] = solsecond(20)
end
p2 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:150)
for i in 1:50
    y[i] = solsecond(50)
end
p3 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:150)
for i in 1:50
    y[i] = solsecond(100)
end
p4 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:150)
for i in 1:50
    y[i] = solsecond(200)
end
p5 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:150)
p6 = plot(1:10,rand(10))
plot(p1,p2,p3,p4,p5,p6layout=(2,3),legend=false)
```

```julia
function solthird(k)
    sum = 0.0
    r = rand(1:length(data_arr),k)
        for i in 1:k
        sum+=data_arr[r[i]]
    end
    sum/k
end 
```

```julia

for i in 1:50
    y[i] = solthird(10)
end
p1 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:140)
for i in 1:50
    y[i] = solthird(20)
end
p2 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:140)
for i in 1:50
    y[i] = solthird(50)
end
p3 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:140)
for i in 1:50
    y[i] = solthird(100)
end
p4 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:140)
for i in 1:50
    y[i] = solthird(200)
end
p5 = plot(x, y,ylims = (100,200),xticks = 0:5:50,yticks=110:10:140)
p6 = plot(1:10,rand(10))
plot(p1,p2,p3,p4,p5,p6layout=(2,3),legend=false)
```
